```python
# Import necessary libraries
from diffusers import StableDiffusionPipeline
import torch
```

**Load a Pre-trained Diffusion Model**
Load a pre-trained model, such as Stable Diffusion, to generate the image:

```python
# Load the Stable Diffusion model
model = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4")

# Move model to CPU
model.to("cpu")
```

```python
# Define your prompt
prompt = "A serene sunset over a mountainous landscape, with a calm river reflecting the colorful sky. The scene includes tall pine trees along the riverbanks, a small wooden cabin on the shore, and vibrant wildflowers scattered across the field. The sky is painted in hues of orange, pink, and purple, with soft clouds floating above the peaks. The atmosphere is peaceful and tranquil, evoking a sense of serenity."

# Generate the image
image = model(prompt).images[0]

# Display the image
image.show()
```

```python
from diffusers import StableDiffusionPipeline
import torch

# Load and move to CPU
model = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4")
model.to("cpu")

# Define prompt and generate image
prompt = "A serene sunset over a mountainous landscape, with a calm river reflecting the colorful sky. The scene includes tall pine trees along the riverbanks, a small wooden cabin on the shore, and vibrant wildflowers scattered across the field. The sky is painted in hues of orange, pink, and purple, with soft clouds floating above the peaks. The atmosphere is peaceful and tranquil, evoking a sense of serenity."
image = model(prompt).images[0]

# Display the image in Colab
image.show()  # This might not work in Colab, so try the next line
image
```
